## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python

# Identify the top 5 clients by total quantity of items ordered
top_clients = df.groupby('client_id')['qty'].sum().nlargest(5).index
#print("top_clients per order line:", df['top_clients'])

# Calculate the total amount spent by each of the top 5 clients
top_clients_spending = df[df['client_id'].isin(top_clients)].groupby('client_id')['total_price'].sum()
#print("top_clients_spending per order line:", df['top_clients_spending'])

# Display the result
print("Total amount spent by each of the top 5 clients by quantity:")
print(top_clients_spending)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

# Identify the top 5 clients by total quantity of items ordered
top_clients = df.groupby('client_id')['qty'].sum().nlargest(5).index

# Calculate the total amount spent by each of the top 5 clients
top_clients_data = df[df['client_id'].isin(top_clients)]

# Summarize data for top 5 clients
summary_df = top_clients_data.groupby('client_id').agg(
    total_units_purchased=pd.NamedAgg(column='qty', aggfunc='sum'),
    total_shipping_price=pd.NamedAgg(column='shipping_price', aggfunc='sum'),
    total_revenue=pd.NamedAgg(column='total_price', aggfunc='sum'),
    total_profit=pd.NamedAgg(column='profit', aggfunc='sum')
).reset_index()

# Display the summary DataFrame
print("Summary DataFrame for the top 5 clients by quantity:")
print(summary_df)
```

```python
# Format the data and rename the columns to names suitable for presentation.
# Define the money columns. 
# Define a function that converts a dollar amount to millions.
# Apply the currency_format_millions function to only the money columns. 
# Rename the columns to reflect the change in the money format. 
# Summarize data for top 5 clients
summary_df = top_clients_data.groupby('client_id').agg(
    total_units_purchased=pd.NamedAgg(column='qty', aggfunc='sum'),
    total_shipping_price=pd.NamedAgg(column='shipping_price', aggfunc='sum'),
    total_revenue=pd.NamedAgg(column='total_price', aggfunc='sum'),
    total_profit=pd.NamedAgg(column='profit', aggfunc='sum')
).reset_index()

# Define the money columns
money_columns = ['total_shipping_price', 'total_revenue', 'total_profit']

# Define a function that converts a dollar amount to millions
def currency_format_millions(amount):
    return f"${amount / 1_000_000:.2f}M"

# Apply the currency format function to only the money columns
for column in money_columns:
    summary_df[column] = summary_df[column].apply(currency_format_millions)

# Rename the columns to reflect the change in the money format
summary_df.rename(columns={
    'total_units_purchased': 'Total Units Purchased',
    'total_shipping_price': 'Total Shipping Price (in millions)',
    'total_revenue': 'Total Revenue (in millions)',
    'total_profit': 'Total Profit (in millions)'
}, inplace=True)

# Display the summary DataFrame
print("Summary DataFrame for the top 5 clients by quantity:")
print(summary_df)
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.

# Extract numeric values from the formatted profit column for sorting
summary_df['Total Profit (in millions)'] = summary_df['Total Profit (in millions)'].replace({'\$': '', 'M': ''}, regex=True).astype(float)

# Sort the updated DataFrame by "Total Profit (in millions)" from highest to lowest
sorted_summary_df = summary_df.sort_values(by='Total Profit (in millions)', ascending=False)

# Reformat the "Total Profit (in millions)" column back to the currency format
sorted_summary_df['Total Profit (in millions)'] = sorted_summary_df['Total Profit (in millions)'].apply(lambda x: f"${x:.2f}M")

# Display the sorted summary DataFrame
print("Sorted Summary DataFrame by Total Profit (in millions) from highest to lowest:")
print(sorted_summary_df)
```
